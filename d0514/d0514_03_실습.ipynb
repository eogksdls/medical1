```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # 한글설정
matplotlib.rcParams['font.size'] = 10 # 글자크기

df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
# 이름과 성적을 가지고 선 그래프을 그리시오

x = df['이름']
y = df['국어']

plt.title('빅데이터 성적그래프')
plt.bar(x,y,label='국어점수',
        color='pink',
        width=0.6)
plt.legend(loc='upper center')
plt.xlabel('학생 이름')
plt.ylabel('국어')
```

```python
# chipotle.tsv 에서 주문중 가장 많이 주문한 메뉴 10개를 꺽은선 그래프로 출력하시오

import pandas as pd
df = pd.read_csv('../z20_data/chipotle.tsv', sep='\t')
df

item_df = df.groupby('item_name').sum()  # index = 'item_name'으로 자동설정
item_df.sort_values('quantity',ascending=False)['quantity'].head(5)
```

```python
# item_name 별 주문한 횟수가 몇번인지 출력하시오
df['item_name'].value_counts()[:5]
```

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # 한글설정
matplotlib.rcParams['font.size'] = 13 # 글자크기

x = df['item_name'].value_counts()[:5].index
y = item_df.sort_values('quantity',ascending=False)['quantity'].head(5).values
y2 = df['item_name'].value_counts()[:5].values


plt.figure(figsize=(10,5))
plt.title('주문량 Top 10',fontdict={'size':18,'weight':600})
plt.xlabel('메뉴',fontdict={'size':12,'weight':600})
plt.ylabel('주문량/주문횟수',fontdict={'size':12,'weight':600})
plt.plot(x,y,label='총 주문량',color='skyblue')
plt.plot(x,y2,label='주문 횟수',color='lightgreen')
plt.xticks(rotation=30)
plt.legend(loc=1)   # 1 = upper right
```
