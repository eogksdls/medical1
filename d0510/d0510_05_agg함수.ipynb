```python
import pandas as pd

df = pd.read_csv('../z20_data/stu_score3.csv',index_col=0)
df
```

```python
# 학교별 국어점수가 가장 높은 점수를 출력하시오
df.groupby(['school','grade'])['kor'].mean()['구로고']
```

```python
# 남자,여자 수를 출력하시오
df.groupby('gender').size()
```

```python
# 합계 컬럼을 추가해서
# 학년별 합계의 평균을 출력하시오
df['total'] = df['kor'] + df['eng'] + df['math']
```

```python
# df.groupby('grade')['total'].mean()  # 성적으로 그룹화 후, total 평균
# df.groupby('grade')['total'].min()  # 성적으로 그룹화 후, total 최솟값
# df.groupby('grade')['total'].max()  # 성적으로 그룹화 후, total 최댓값
# df.groupby('grade')['total'].sum()  # 성적으로 그룹화 후, total 합

# apply 함수 - 함수호출, agg - 기본함수호출
df.groupby(['school','grade'])['total'].agg(['mean','max','min','sum'])

# 기본함수가 mean, max, min, sum 이 있다
```
