```python
import pandas as pd

df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
# 합계라는 컬럼을 생성하고
df['합계'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
# 3학년중에 성적이 가장 낮은 학생을 출력하시오
df['학년'] == 3  # 3학년 학생들만 추출
min_3_score = df.groupby('학년').get_group(3)['합계'].min()
# df.loc['합계',df.groupby('학년').get_group(3)['합계'].min()]  # 학년별로 성적이 가장 낮은 사람
 
```

```python
# 3학년 중에 합계가 가장 낮은 점수를 찾아서, 합계점수 조건을 가지고 row 데이터 겁색
df[df['합계'] == min_3_score]
```

```python
### 평균 컬럼을 추가해서 grade 컬럼에 A >= 90, B >= 80, C는 70, D는 60점 이상, 나머지 F

df['평균'] = df['합계'] / 5.0
```

```python
def grade(score):
    if score >= 90 : return 'A'
    elif score >= 80 : return 'B'
    elif score >= 70 :  return 'C'
    elif score >= 60 : return 'D'
    else : return 'F'
```

```python
df['grade'] = df['평균'].apply(grade)
df
```

```python
# 학년 간 비율
(df['학년'].value_counts(normalize=True)*100).astype(str)+'%'
```

```python
# SW특기 컬럼을 모두 대문자로 변경, NaN데이터는 모두 python으로 입력
def upper_change(text):
    # 
    if pd.isnull(text) : 
        return "python".upper()
    else:
        return text.upper()
```

```python
# 또는 df.fillna('PYTHON', inplace=True) 처리한 후 upper()만 해줘도 됨
```

```python
df['SW특기'] = df['SW특기'].apply(upper_change)
df
```
