```python
import pandas as pd

df = pd.read_csv('../z20_data/drinks.csv')
df
```

```python
# 어떤 값들이 int 처리되어 있는지 확인 -> 주로 다루는 데이터가 int일 경우가 많기 때문에
# 확인하는 것은 필수!!!
df.info()
```

```python
# NA -> NN 으로 바꾸기
df.fillna('NN',inplace=True)
df
```

```python
len(df[df['continent'].isnull()]) 
# 결측치 0개!
```

```python
# 술이 팔린 개수를 더한 컬럼을 추가하여 제일 많이 팔린 대륙 호출
df['total'] = df['beer_servings'] + df['spirit_servings'] + df['wine_servings']
df
```

```python
# 대륙별로 최소,최대,평균,합계를 구하시오
# continent
# 대륙별로 최소값
# df.groupby('continent')['total'].min()
# df.groupby('continent')['total'].max()
# df.groupby('continent')['total'].mean()
# df.groupby('continent')['total'].sum()

df.groupby('continent')['total'].agg(['count','min','max','mean','sum'])
```

```python
df.groupby('continent')[['beer_servings','spirit_servings','wine_servings']].agg(['count','min','max','mean','sum'])
```

```python
# 전체 대륙 평균 맥주소비량
df['beer_servings'].mean()
# 약 106병 정도
```

```python
# 전체 평균보다 높은 맥주소비량을 가지는 대륙을 출력하시오

# 1. 전체 평균 소비량
beer_mean = df['beer_servings'].mean()
# 2. 대륙별 평균 소비량
continent_over_mean = df.groupby('continent')['beer_servings'].mean()
continent_over_mean[continent_over_mean >= beer_mean]
```

```python
# total_litres_of_pure_alcohol이 전체 평균보다 높은 대륙을 출력
alcohol_mean = df['total_litres_of_pure_alcohol'].mean()
alcohol_mean
# 약 4.71

continent_alcohol_mean = df.groupby('continent')['total_litres_of_pure_alcohol'].mean()
continent_alcohol_mean[continent_alcohol_mean >= alcohol_mean]
```
