```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # 한글설정
matplotlib.rcParams['font.size'] = 13 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False   # 마이너스 부호 넣기
```

```python
x = [1,2,3]
y = [2,4,8]

# 그래프 그리기, 선두께 = linewidth, label = 범례, alpha =  투명도, linestyle = 선스타일
plt.plot(x,y,linewidth=3,marker='X',ms=7,mec='black',
         mfc='yellow',label='국어',alpha=0.7,ls=':')
# plt.legend(loc='lower right')
# 범례 위치 소수점 설정(원하는 위치로 설정할 수 있음)
plt.legend(loc=(0.2,0.5))
plt.title('막대그래프')
# x,y축 표시 간격을 조정
plt.yticks([5,7,9,11,13,15])
plt.xticks([1,2,3])
```
