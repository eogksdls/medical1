```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 11
matplotlib.rcParams['axes.unicode_minus'] = False
import pandas as pd
import numpy as np
```

```python
values = [30,25,20,13,10,2]  # 100
title = ['Python','Java','Javascript','c#','c','ETC']
explode = [0.1]*6
colors = ['#AF48F5','#3D9CF5','#F5C53D','#F58B3D','#F55C4D','#F4A2A8']
wedgeprops = {'width':0.6, 'edgecolor':'w','linewidth':2}


# values : 값
# labels : 표시(글자)
# autopct : 파이내 글자표시 및 퍼센트표시
#pctdistance : 파이내 글자 표시 위치
# startangle : 원형이 시작하는 위치
# colors : 그래프 색상
# counterclock : 그래프 내 데이터 배치 - 시계방향, 반시계방향
# wedgeprops :  도넛형태, 그래프 테두리 색상, 라인두께 설정 가능
# fontdic : 글자 크기, 두께 설정 가능
```

```python
plt.pie(values, labels=title, autopct='%1.f%%', startangle=90, explode=explode, colors=colors, counterclock=False,\
    wedgeprops=wedgeprops)
plt.title('전체 언어별 선호도',fontdict={'size':15,'weight':600})
plt.show()
```

```python
plt.pie(values, labels=title, autopct='%1.f%%', startangle=90, colors=colors, counterclock=False,\
    wedgeprops=wedgeprops)
plt.title('전체 언어별 선호도',fontdict={'size':15,'weight':600})
plt.legend(loc=(1.0,0.8),title='언어별 선호도')
plt.show()
```
