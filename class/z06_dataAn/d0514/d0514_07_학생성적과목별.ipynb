```python
# score.xlsx 합계 컬럼을 생성해서 막대 그래프와 가로 막대그래프를 생성하시오
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # 한글설정
matplotlib.rcParams['font.size'] = 10 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False   # 마이너스 부호

import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df.head(3)
```

```python
x = df.head(5)['이름']
y = df.head(5)['국어']
y2 = df.head(5)['영어']
y3 = df.head(5)['수학']
y4 = df.head(5)['과학']
y5 = df.head(5)['사회']
```

```python
plt.plot(x,y,label='국어',marker='o')
plt.plot(x,y2,label='영어',marker='D',ls='--')
plt.plot(x,y3,label='수학',marker='s',ls='-.')
# plt.plot(x,y4,label='과학')
# plt.plot(x,y5,label='사회')
for i, txt in enumerate(y3):
    # 수학만
    plt.text(x[i],y3[i]+2,txt,ha='center',color='black')    
plt.title('학생성적그래프',fontdict={'size':13,'weight':600})
plt.legend(loc='upper center',ncol=3)
plt.figure(figsize=(30,10))
```

```python
colors = ['bisque','tan','orange','darkkhaki','olive']
bars = plt.bar(x,y,label='국어',color=colors, width=0.6, alpha=0.5)
# 패턴넣기
bars[0].set_hatch('/') 
bars[1].set_hatch('x') 
bars[2].set_hatch('+') 
bars[3].set_hatch('|') 
bars[4].set_hatch('o') 
plt.ylim(40,100)
for i, txt in enumerate(y):
    plt.text(x[i],y[i]+2,txt,ha='center',color='black')  
    
plt.title('학생성적그래프',fontdict={'size':13,'weight':600})
plt.legend(loc='upper right',ncol=3)
plt.figure(figsize=(30,10))

plt.show()
```
