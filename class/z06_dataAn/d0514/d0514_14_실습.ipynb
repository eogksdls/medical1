```python
# 막대그래프 score.xlsx 합계 점수를 뽑아서 그래프로 표현하시오
import pandas as pd

df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
df['합계'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 11
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = df['이름']
y = df['합계']

plt.bar(x,y,label='전체성적',color='pink')
plt.title('학생 성적 그래프', fontdict={'size':14,'weight':600})
plt.grid(axis='y', alpha=0.3, linestyle='--', linewidth=2)
for i, txt in enumerate(y):
    plt.text(x[i],y[i]+1.5,txt,ha='center',color='red')
plt.legend(loc='upper center')
plt.ylim(250,500)
```

```python
plt.barh(x,y,label='전체성적',color='pink')
plt.title('학생 성적 그래프', fontdict={'size':14,'weight':600})
plt.grid(axis='x', alpha=0.3, linestyle='--', linewidth=2)
plt.legend(loc='upper center')
plt.xlim(250,500)
for i, txt in enumerate(y):
    plt.text(y[i]+8,x[i],txt,ha='center',color='red')
```
