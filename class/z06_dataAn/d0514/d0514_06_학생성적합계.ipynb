```python
# score.xlsx 합계 컬럼을 생성해서 막대 그래프와 가로 막대그래프를 생성하시오
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx')

df['합계'] = df['국어'] + df['영어'] + df['수학'] + df['사회'] + df['과학']
df
```

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # 한글설정
matplotlib.rcParams['font.size'] = 13 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False   # 마이너스 부호
```

```python
x = df['이름']
y = df['합계']

plt.figure(figsize=(13,5))
plt.title('학생 성적 그래프',fontdict={'size':18,'weight':600})
plt.bar(x,y,label='성적', color='lightgreen')
plt.legend(loc=1)
plt.ylim(250,500)
for i, txt in enumerate(y):
    plt.text(x[i],y[i]+3,txt,ha='center',color='blue')
plt.xlabel('학생 이름',fontdict={'size':13,'weight':600})
plt.ylabel('학생 성적',fontdict={'size':13,'weight':600})
```

```python
plt.figure(figsize=(13,5))
plt.title('학생 성적 그래프',fontdict={'size':18,'weight':600})
plt.barh(x,y,label='성적', color='skyblue')
plt.legend(loc=1)
plt.xlim(250,500)
for i, txt in enumerate(y):
    plt.text(y[i]+5,x[i],txt,ha='center',color='blue')
plt.xlabel('학생 성적',fontdict={'size':13,'weight':600})
plt.ylabel('학생 이름',fontdict={'size':13,'weight':600})
```
