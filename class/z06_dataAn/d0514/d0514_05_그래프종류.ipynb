```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # 한글설정
matplotlib.rcParams['font.size'] = 13 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False   # 마이너스 부호 넣기

x = [1,2,3]
y = [2,4,8]
```

```python

# 라인 스타일
plt.plot(x,y,label='국어',linestyle='--',marker='d')
plt.legend(loc='upper center')

# 좌표에 데이터 2,4,8의 값을 출력
# plt.text()
# x = [1,2,3]
# y = [2,4,8] : 데이터값
for i, txt in enumerate(y):
    plt.text(x[i],y[i]+0.2,txt,ha='center',color='blue')
plt.yticks([x for x in range(0,11)])


# 보여지는 그래프 전체크기가 커짐
plt.figure(figsize=(10,4))  # 가로세로 조절 가능
plt.figure(dpi=200)  # 전체크기, 해상도
# plt.savefig('g.png',dpi=100)  # figure 저장, dpi 이미지 크기 up, 해상도도 좋아짐


plt.show()
```

```python
plt.bar(x,y,label='국어',width=0.6, color='lightgreen')
plt.legend(loc='upper center')

for i, txt in enumerate(y):
    plt.text(x[i],y[i]+0.2,txt,ha='center',color='blue')
plt.yticks([x for x in range(0,11)])

plt.show()
```

```python
import pandas as pd

df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df.head(3)
```

```python
# 키를 가지고 키 비교 꺽은선 그래프를 출력하시오
student = df['이름']
height = df['키']

plt.plot(student,height,label='키',ls='--',marker='*',mfc='black',mec='black',color='orange',
         linewidth=2)
plt.legend(loc='upper center')
plt.title('학생 키 그래프',fontdict={'weight':600})
plt.xlabel('학생이름',fontdict={'size':10,'weight':600})
plt.ylabel('키',fontdict={'size':10,'weight':600})

for i, txt in enumerate(height):
    plt.text(student[i],height[i]+1,txt,ha='center',color='black',fontdict={'size':10})
# plt.yticks([160,170,180,190,200,210])   # y축 눈금표시
plt.ylim(165,205)       # y축 눈금 표시 : 범위만 지정해주면 알아서 눈금 설정

plt.figure(figsize=(13,5))
plt.show()

```

```python
# 막대그래프( bar )
plt.bar(student,height,label='키',color='orange')
plt.legend(loc='upper center')
plt.title('학생 키 그래프',fontdict={'weight':600})
plt.xlabel('학생이름',fontdict={'size':10,'weight':600})
plt.ylabel('키',fontdict={'size':10,'weight':600})

for i, txt in enumerate(height):
    plt.text(student[i],height[i]+1,txt,ha='center',color='blue',fontdict={'size':10})
# plt.yticks([160,170,180,190,200,210])   # y축 눈금표시
plt.ylim(165,205)       # y축 눈금 표시 : 범위만 지정해주면 알아서 눈금 설정

plt.figure(figsize=(13,5))
plt.show()
```

```python
# 누운 막대 그래프 ( barh ) -> x축에 데이터값 표시됨
plt.barh(student,height,label='키',color='orange')
plt.legend(loc='upper right')
plt.title('학생 키 그래프',fontdict={'weight':600})
plt.xlabel('키',fontdict={'size':10,'weight':600})
plt.ylabel('학생이름',fontdict={'size':10,'weight':600})

for i, txt in enumerate(height):
    plt.text(height[i]+1.8,student[i],txt,ha='center',color='blue',fontdict={'size':10})
    
plt.xlim(165,205)       # 눈금 표시 : 범위만 지정해주면 알아서 눈금 설정
plt.figure(figsize=(13,5))
plt.show()
```
