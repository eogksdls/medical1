### 데이터 선택 - 조건

```python
# pandas 조건문 : 컬럼의 조건문을 만족하는 행
filt = df['키'] > 190
df[filt]  # 키가 190 이상인 학생만 출력
```

```python
df['키'] > 190
df[df['키']>190]
```

```python
# 키가 평균보다 큰 학생의 키값만 출력하시오
# 1. 평균 구하기
df['키'].mean()

# 2. 조건식
filt = df['키'] > df['키'].mean()

# 3. 완료
df.loc[filt,['이름','키']]

```

```python
## 국어 85점 이상, 영어 85점 이상 이름,학교,국어,영어 출력하시오

# 1. 국어,영어 85점 이상
df['국어'] >= 85
df['영어'] >= 85

# # 2. 조건식 쓰기
filt = (df['국어'] >= 85) & (df['영어'] >= 85)


df.loc[filt,['이름','학교','국어','영어']]
df.loc[~filt,['이름','학교','국어','영어']] # else 국어나 영어성적 중 85점 미만 or 개념으로 바뀜
```
