```python
import pandas as pd
df = pd.read_csv('stu_score1.csv',index_col='no')
df
```

#### 1~5번까지 kor,eng,math만 출력하시오

```python
# df.loc[인덱스 이름]

df.loc[[1,2,3,4,5],['kor','eng','math']]
df.loc[1:5,'kor':'math']
```

### iloc[인덱스위치] 출력

```python
# df.loc[인덱스 위치]
df.iloc[0]  # df.loc['1번'] 같음
df.iloc[3] # 4번째 row 데이터 출력
df.iloc[2:4] # 3번째부터 4번째 row데이터 출력
df.iloc[[2,5,6]]  # 순서가 0부터 시작해서 하나씩 밀리는 것

# 3,6,7번 학생의 국어성적(column=2) 출력
df.iloc[[2,5,6],2]

# 5,8번 학생의 성적 출력
df.iloc[[4,7]]
```

```python
df = pd.read_csv('score.csv',index_col='지원번호')
df

# 5번, 8번 학생의 성적 출력
df.iloc[[4,7]]
# df.loc[['5번','8번']]

# 4번 학생 키 출력
df.iloc[3,2]
# df.loc['4번','키']

# 4,6번 학생의 국어, 사회를 출력하시오
df.iloc[[3,5],[3,7]]
# df.loc[['4번','6번'],['국어','사회']]

# 마지막 row 출력
df.iloc[-1]
df.iloc[-4:]

df.iloc[2:4]
df.iloc[2:4,0:5]

# 마지막 column 출력
# df[df.columns[-1]]
```
