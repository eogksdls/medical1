### skiprows 상단부분 지정한 만큼 제거 후 읽어오기

```python
df = pd.read_csv('stu.csv',skiprows=1)  # 1번째 행은 건너뛴 채 출력

df.set_index('1번',inplace=True)
df
```

```python

df = pd.read_csv('stu.csv',nrows=4, skiprows=1)

df

# 파일저장
# df.to_csv('stu2.csv',encoding='utf-8-sig')
```
