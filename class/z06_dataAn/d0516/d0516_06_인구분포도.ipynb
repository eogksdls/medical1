```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 11
matplotlib.rcParams['axes.unicode_minus'] = False
import pandas as pd
import numpy as np
```

연령별 인구현황

1. 202301 연령별 인구수를 가로 막대그래프로 출력하시오.
2. 남성, 백의 자리 수 이하 버림

```python
df_m = pd.read_excel('../z20_data/20230104_연령별인구현황2.xlsx',index_col='행정기관',skiprows=3,nrows=1,usecols='B,E:Y',thousands=',')
df_f = pd.read_excel('../z20_data/20230104_연령별인구현황2.xlsx',index_col='행정기관',skiprows=3,nrows=1,usecols='B,AB:AV',thousands=',')
# df.values//100*100
# df.values.str.replace(',','').astype(int)
```

```python
x = df_m.columns
y1 = -df_m.values//100*100
y2 = df_f.values//100*100
# x
# y[0]

# 가로 막대 그래프 그리기
plt.figure(figsize=(8,7))
plt.barh(x,y1[0],label='남자',color='#A1ABE4')
plt.barh(x,y2[0],label='여자',color='#E3A0B6')
plt.legend(loc='upper center',ncol=2)
# x축 범위 설정
plt.xlim(-2500000,2500000)
# 지수 표현방식 없애기
current_values = plt.gca().get_xticks()
plt.gca().set_xticklabels(['{:,.0f}'.format(x) for x in current_values])

plt.show()
```
