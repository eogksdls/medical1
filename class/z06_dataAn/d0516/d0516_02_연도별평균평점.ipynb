```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 11
matplotlib.rcParams['axes.unicode_minus'] = False
import pandas as pd
import numpy as np
```

```python
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
```

대한민국 영화 중에서 관객 수가 가장 많은 상위 8개의 데이터
1. 개봉연도별 평점 선그래프를 출력하시오
- 선은 red, marker 표시
- x축 눈금 표시 : 2005,2010,2015,2020
- y축 눈금범위 7,10

```python
df = pd.DataFrame(data)
df
```

```python
rank_mean = df.groupby('개봉 연도')['평점'].mean()
# rank_mean.values
```

```python

# 그래프 가로,세로 크기 설정
plt.figure(figsize=(10,5))
# 그래프 그리기
plt.plot(rank_mean.index,rank_mean.values,color='red',marker='*',ms=8,mec='blue',mfc='blue',ls='--')
# x,y축 설정
# plt.xticks([2005,2010,2015,2020])
plt.xlim(2005,2020)
plt.ylim(7,10)
# 제목설정
plt.title('개봉 연도별 평균 평점',fontdict={'weight':600})
# x,y축 범례
font = {'size':10,'weight':600}
plt.xlabel('개봉 연도',fontdict=font)
plt.ylabel('평균 평점',fontdict=font)
# grid 
plt.grid(ls='--',alpha=0.5)
# 범례값 출력
for i, txt in enumerate(rank_mean.values):
    plt.text(rank_mean.index[i],rank_mean.values[i]+0.1,txt,ha='center',color='blue')
```
