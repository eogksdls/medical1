```python
df = pd.read_csv('../z20_data/picher_stats_2017.csv')
df.head(3)
```

```python
label = df['팀명']
label.unique()
```

```python
label_numpy = label.values.reshape(-1,1)
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False)
ohe.fit(label_numpy)
```

```python
ohe.categories_
```

```python
label_ohe = ohe.transform(label_numpy)
label_ohe
```

```python
ohe_df = pd.DataFrame(label_ohe, columns=ohe.categories_)
ohe_df
```
