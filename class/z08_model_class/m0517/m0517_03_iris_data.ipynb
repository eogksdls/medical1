```python
from sklearn import svm,metrics
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import requests
# maplotlib.rcParams['font.family'] = 'AppleGothic Gothic   # Mac 사용자
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('../z20_data/02_iris.csv')
df.head(3)
```

1. 데이터 전처리 : train_data, test_data

```python
### rows = 100개를 출력하시오
# df[df.columns[:-1]].head(100)
train_input = df.iloc[:100,:-1]

### rows = 50개를 출력하시오
test_input = df.iloc[100:,:-1]
```

```python
### train_target = rows 100개 출력
train_target = df['Name'][:100]

### test_target = rows 50개 출력
test_target = df['Name'][100:]
```

2. 데이터 학습시키기 

```python
clf = svm.SVC()
# 훈련
clf.fit(train_input,train_target)
```

3. 데이터 예측하기 : 붓꽃 종류 맞추기

```python
result = clf.predict(test_input)
# result
```

4. 정답률 구하기

```python
score = metrics.accuracy_score(test_target,result)   # 예측하고자 했던 데이터의 결과값, 예측한 값과의 비교
print("정답률 : ",score)
```

```python
# index로 df에 대입해서 가져오면, 랜덤으로 데이터가 섞여서 가져올 수 있음
# train_input = df.iloc[index[:100],:-1]
# train_target = df.iloc[index[:100],4]
# test_input = df.iloc[index[100:],:-1]
# test_target = df.iloc[index[100:],4]
```

```python
# 1. 데이터 전처리
datas_input = df.iloc[:,:-1]
datas_target = df.iloc[:,4]

# 모든 data를 가지고, train,test 데이터 분리
# train, test 데이터를 자동으로 분리해줌
train_input,test_input,train_target,test_target = train_test_split(datas_input,datas_target)
```

```python
# 2. 데이터 학습시키기
clf = svm.SVC()
# 훈련
clf.fit(train_input,train_target)

# 3. 예측
result = clf.predict(test_input)

# 4. 정답률 
# score = metrics.accuracy_score(test_target,result)
score = clf.score(test_input,test_target)
print("정답률 : ",score)
```
