```python
import requests
import time
import random
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

# 화면이 나타나는 것을 확인
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pyautogui
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36","Accep-Language":"ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7"}
url = "https://namu.wiki/RecentChanges"
```

```python
# 브라우저 열기
browser = webdriver.Chrome()
browser.maximize_window()
browser.get(url)
```

```python
# 데이터 불러오기 - 파일저장
soup = BeautifulSoup(browser.page_source,"lxml")
with open("test.html","w",encoding="utf8") as f:
    f.write(soup.prettify())
    # 표부분 전체 위치 값 찾기
    section = soup.find("div",{"class":"_05sl5euW"})
```

```python
lists = section.find_all("div",{"class":"RCuwAzkH"}) #100개
```

```python
urls = []
for i in range(1,len(lists)):
    urls.append("https://namu.wiki"+lists[i].a['href'])
print("완료")
```

```python
urls
```

상세페이지 내 title,category,content

```python
title = []
category = []
content = []
for i in range(len(lists)):
    try:
        # 상세페이지 title,category, 
        url = urls[i]
        print('순번 : ',i)
        print("url : ",url)
        res = requests.get(url)
        res.raise_for_status()
        # 데이터 불러오기
        soup = BeautifulSoup(res.text, 'lxml')
        # title
        title.append(soup.find("a",{"class":"Wu5cnaNH"}).text)
        # category
        category.append(soup.find("ul",{"class":"cvIQfS6I"}).text)
        # content
        contents = soup.find("div",{"class":"_0NDK1MJM"})
        cont = contents.find("span",{"id":"개요"}).parent.nextSibling.text
        content.append(cont)
    except:
        print("에러처리 : ",i)
```

```python
data = {"title":title,"category":category,"content":content}
```

```python
import pandas as pd
# content의 개수와 title&category 개수와 달라서 Dataframe이 생성되지 않음
df = pd.DataFrame.from_dict(data, orient='index') 
df = df.T  # 열과 행 바꿔주긔
df
```
